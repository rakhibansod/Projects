##POPULARITY BASED RECOMMENDATIONS SYSTEM


```python
ratings.merge(books,on='ISBN')
```

```python
ratings.merge(books,on='ISBN').shape
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
ratings_with_name.groupby('Book-Title').count()
```

```python
num_ratings_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns={'Book-Rating':'num-ratings'},inplace=True)
num_ratings_df
```

```python
avg_ratings_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings_df.rename(columns={'Book-Rating':'avg-ratings'},inplace=True)
avg_ratings_df
```

```python
popular_df = num_ratings_df.merge(avg_ratings_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num-ratings']>=250].sort_values('avg-ratings',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on ='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num-ratings','avg-ratings']]
```

```python
popular_df
```

## Collaborative Filtering Based Recommender

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
hoshiyar_bacche = x[x].index
```

```python
ratings_with_name['User-ID'].isin(hoshiyar_bacche)
```

```python
ratings_with_name
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(hoshiyar_bacche)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
finals_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
pt = finals_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def recommend(book_name):
    # Check if the book_name is in the index
    if book_name not in pt.index:
        print(f"The book '{book_name}' is not in the dataset.")
        return
    
    # Index fetch
    index = np.where(pt.index == book_name)[0][0]
    
    # Assuming 'similarity_score' is a 2D array or DataFrame
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in Bottle')
```
